```python
new_df = pd.DataFrame()
```

```python
for x in range(1,397):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=ad5f333b4cccb1d8cbe21de5d526208a&language=en-US&page={}'.format(x))
  df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  new_df = pd.concat([new_df,df],ignore_index=True)


```
