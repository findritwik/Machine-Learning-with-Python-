```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import MissingIndicator,SimpleImputer
df = pd.read_csv('/content/train (3).csv',usecols=['Age','Fare','Survived'])
```

```python
df
```

```python
df.info()
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
si = SimpleImputer()
x_train1 = si.fit_transform(x_train)
x_test1 = si.transform(x_test)
```

```python
x_train1 = pd.DataFrame(x_train1)
x_test1 = pd.DataFrame(x_test1)
```

```python
x_train1
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train1,y_train)

y_pred = clf.predict(x_test1)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
mi = MissingIndicator()
mi.fit(x_train)

x_train_missing = mi.transform(x_train)
x_test_missing = mi.transform(x_test)
```

```python
x_train_missing
```

```python
si1 = SimpleImputer(add_indicator=True)
x_train_trf = si1.fit_transform(x_train)
x_test_trf = si1.transform(x_test)
```

```python
x_train_trf = pd.DataFrame(x_train_trf)
x_test_trf = pd.DataFrame(x_test_trf)
```

```python
x_train1
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train_trf,y_train)

y_pred = clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
